```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('home.csv')
df.head()
```

```python
df_dummies = pd.get_dummies(df['town'])
df_dummies
```

```python
df_dummies= pd.concat([df_dummies,df],axis=1)
df_dummies
```

```python
df_dummies.drop(['monroe township','town'],axis=1,inplace=True)
```

```python
df_dummies
```

```python
x = df_dummies.iloc[:,:3].values
x
```

```python
y = df_dummies.iloc[:,3].values
y
```

```python
lin = LinearRegression()
lin.fit(x,y)
```

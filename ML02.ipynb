```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph')
plt.show()
```

```python
time = np.arange(0,2.0,0.01)
func = np.cos(2*np.pi*time)
plt.plot(time,func,'r')
plt.grid()
plt.xlabel('time')
plt.ylabel('value')
plt.title('cosine graph')
plt.show()
```

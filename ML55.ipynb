```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data = pd.read_csv('startup.csv')
data.head()
```

```python
real_x = data.iloc[:,0:4].values
real_y = data.iloc[:,4].values
#real_y = real_y.reshape(-1,1)
real_x
```

```python
le = LabelEncoder()
real_x[:,3] = le.fit_transform(real_x[:,3])
ohe = OneHotEncoder(categorical_features=[3])
real_x = ohe.fit_transform(real_x).toarray()
real_x
```

```python
training_x,testing_x,training_y,testing_y = train_test_split(real_x,real_y,test_size=0.2,random_state=0)
training_x
```

```python
lin = LinearRegression()
lin.fit(training_x,training_y)
```

```python
lin.predict(testing_x)
```

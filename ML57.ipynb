```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.compose import ColumnTransformer
```

```python
clt = ColumnTransformer([('step',OneHotEncoder(),[3])],remainder="passthrough")
```

```python
df = pd.read_csv('startup.csv')
df.head()
```

```python
real_x = df.iloc[:,0:4].values
real_y = df.iloc[:,4].values
```

```python
le = LabelEncoder()
real_x[:,3] = le.fit_transform(real_x[:,3])
oneHE = OneHotEncoder(categorical_features=[3])
real_x = oneHE.fit_transform(real_x).toarray()
real_x
```

```python
real_x = real_x[:,1:]
real_x
```

```python
training_x,testing_x,training_y ,testing_y = train_test_split(real_x,real_y,test_size=0.2,random_state=0)
```

```python
lin = LinearRegression()
lin.fit(training_x,training_y)
```

```python
lin.predict(testing_x)
```

```python
lin.coef_
```

```python
lin.intercept_
```
